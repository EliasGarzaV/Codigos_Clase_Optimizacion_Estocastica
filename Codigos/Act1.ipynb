## Problema 4

Lo que sucede es que en ningun momento dicen que el software detecta fraudes el 90% de las veces, en cambio lo que dicen es que detecta el 90% de los fraudes por lo que la probabilidad real de que el software detecte es condicional a que exista algo. Formalmente, si $E$ es que exista fraude y $D$ es que se detecte tenemos que 

$$P(D \mid E) = 90\%$$ 

Entonces si la probabilidad de que exista fraude es $p$ (Osea $P(E) = p$) entonces nos queda que

$$P(D) =P(E) \cdot P(D \mid E) = 0.9p$$

Esto nos dice que la probabilidad de que el programa arroje un resultado positivo es $0.9p$ lo cual puede ser bastante bajo dependiendo de $p$ por lo que que no se haya detectado nada no es motivo para preocuparse.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import statistics 

fig, ax = plt.subplots(2, figsize=(15,10))
ax[0].set_title('Primer Momento')
ax[1].set_title('Segundo Momento')

population = np.random.normal(750, 100, size = 10000)

trayectorias = []
for i in range(5):
    means = []
    vars = []
    for j in range(50, 5500, 50):
        sample = np.random.choice(population, j)
        vars.append(statistics.variance(sample))
        means.append(sum(sample) / j)
    
    ax[0].plot(list(range(50, 5500, 50)), means)
    ax[1].plot(list(range(50, 5500, 50)), vars)
```

```python
def confidence_interval_var(s, n, conf):
    part1 = ((n - 1) * s * s) / (stats.chi2.ppf(conf / 2, n - 1))
    part2 = ((n - 1) * s * s) / (stats.chi2.ppf(1 - (conf / 2), n - 1))
    return (min(part1, part2), max(part1, part2))

confidence_interval_var(100, 500, 0.97)
```
